**Скачайте файл с данными посещаемости сайта Яндекс.Метрика.
При помощи функции np.where создайте серию, в которую будет записываться 1, 
если посещения были сделаны роботами с браузера, содержащего Chrome в названии, и 0 в других случаях. 
Поместите получившуюся серию в переменную chrome_robots.**

```python
ya_df = pd.read_csv('7_browser.csv')
```

```python
ya_df.head()
```

```python
ya_df['chrome_robots'] = np.where((ya_df.user_type.str.contains('Robots') & ya_df.browser.str.contains('Chrome')), 1, 0)
```

```python
chrome_robots = ya_df.chrome_robots
```
