**Дополнительный проект
В данном проекте вы поработаете с довольно нетипичными для аналитика данными: gff файлом, который описывает геном мухи.**

```python
import pandas as pd
```

```python
fly_df = pd.read_table('gff.tsv')
```

```python
fly_df
```

**Какое значение атрибута Parent самое частое?**

```python
fly_df.attributes.str.extract(r'Parent=([\w\.]+)').value_counts(normalize=True).mul(100).round()
# извелкаем колонку по паттерну, () для экстракта обязательны \w буква \. точка + все слова
```

**Какое самое частое значение в колонке type?**

```python
fly_df.type.str.extract(r'(\w+)').value_counts(normalize=True).mul(100).round()
```
